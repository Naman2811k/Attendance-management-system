```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("ocean.webp",0)
```

```python
he_img = cv2.equalizeHist(img)
```

```python
plt.figure(figsize=(10,4))
```

```python
plt.subplot(1,3,1);
plt.imshow(he_img, cmap='gray');
plt.title("histogram_Equalization");
plt.axis('off');
```

```python
cv2.imwrite('hist_equalized.png', he_img)
```

```python
min_val,max_val=np.min(img),np.max(img)
```

```python
cs_img=((img-min_val)/ (max_val-min_val)*255).astype(np.uint8)
```

```python
plt.subplot(1,3,2);
plt.imshow(cs_img,cmap='gray');
plt.title("contrast_Stretching");
plt.axis('off');
```

```python
cv2.imwrite('contrast_stretched.png',cs_img)
```
